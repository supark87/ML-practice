```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generate artificial data with scikit
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,
                 cluster_std=1.8,random_state=101)
```

```python
data[0].shape
#200 samples 2 features
```

```python
#plot all the rows, first column vs all the rows, second column
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
#cluster 
data[1]
```

```python
from sklearn.cluster import KMeans
#unsupervised
#step1 : randomly items to cluster/ reassign centroid
#step2 : calculate new centroid until 
#cluster variation cannot be reduced any further
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ ##predicted labels
```

```python
data[1] ##correct labels
```

```python
fig , (ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
